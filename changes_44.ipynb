# Changes in the 4.4 Release

## Jupyter Notebook 5.0

See http://jupyter-notebook.readthedocs.io/en/latest/changelog.html#release-5-0-0

## Many Updated Packages

* [ipywidgets](https://ipywidgets.readthedocs.io/en/stable/) Updated to 7.0.

## New Packages

* [tensorflow](https://www.tensorflow.org/)
* [ipyvolume](http://ipyvolume.readthedocs.io/en/latest/index.html)
* [py3dmol](https://3dmol.csb.pitt.edu/)
* [K3D Jupyter](https://github.com/K3D-tools/K3D-jupyter). New 3D visualization.
