# Plotting

## Matplotlib

By default matplotlib plots are static inline images.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
xp = np.random.rand(N)
yp = np.random.rand(N)
colors = np.random.rand(N)
radii = 15 * np.random.rand(N)
area = np.pi * radii**2  # 0 to 15 point radiuses

plt.scatter(xp, yp, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
x = np.linspace(0, 10)
line, = plt.plot(x, np.sin(x), linewidth=2)
plt.show()
```

You can switch to a more interactive output for the notebook by using

## Seaborn

seaborn is a library built on top of matplotlib.  It has excellent statistical plots. It also has a different style, which you can easily use for more basic plots.

```python
import seaborn as sns
sns.set_style("darkgrid")

plt.figure()
line, = plt.plot(x, np.sin(x), linewidth=2)
plt.show()
```

```python
plt.figure()
plt.scatter(xp, yp, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
from bokeh.plotting import figure, show, output_file, vplot
from bokeh.io import output_notebook
```

```python
# this needs to be alone in the cell
output_notebook()
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p = figure(title="Basic Line Example", tools=TOOLS)
p.line(x, np.sin(x), legend="sin(x)")

show(p);
```

```python
# we will color this by radius
import matplotlib as mpl
bcolors = [
    "#%02x%02x%02x" % (int(r), int(g), int(b)) for r, g, b, _ in 255*mpl.cm.jet(mpl.colors.Normalize()(radii))
]
p = figure(title="Basic Scatter Plot", tools=TOOLS)
p.circle(xp, yp, radius=radii/500, fill_color=bcolors, fill_alpha=0.5, line_color=None)
show(p);
```
