## Installed Modules

This notebook is running in a linux environment with a large collection of hundreds of scientific packages based on Continuum Analytics Anaconda distribution.  You can get a list of installed packages from within a notebook by executing the cell below.

If we just do 

    !conda list

the notebook server will run the command and return the result,  Unfortunately that will just give us the packages installed in the server's environment.  We really want the results for the packages in the kernel we are using.  To get that we have to do something like:

````python
# Use internal APIs to get list of packages in this environment
# WARNING: using undocumented API calls
from conda.cli.main_list import print_packages
from conda.config import root_dir
print_packages(root_dir, piplist=True)
````
