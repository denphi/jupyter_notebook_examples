```python
from nanohubuidl.ipywidgets import buildJSX
gingerbread = buildJSX(
    '''<svg class="gingerbread" width="200" height="200" viewBox="-100 -100 200 200">
  <circle class="body" cx="0" cy="-50" r="30" />

  <circle class="eye" cx="-12" cy="-55" r="{left_eye:integer(3)}" />
  <circle class="eye" cx="12" cy="-55" r="{right_eye:integer(3)}" />
  <rect class="mouth" x="-10" y="-40" width="20" height="5" rx="2" />

  <line class="limb" x1="-40" y1="-10" x2="40" y2="-10" />
  <line class="limb" x1="-25" y1="50" x2="0" y2="-15" />
  <line class="limb" x1="25" y1="50" x2="0" y2="-15" />

  <circle class="button" cx="0" cy="-10" r="5" />
  <circle class="button" cx="0" cy="10" r="5" />
</svg>''')

gingerbread(left_eye=5)
```

```python
import ipywidgets as w
a = gingerbread(left_eye=10)
b = gingerbread(right_eye=10)
w.HBox([a,b])
```

```python
a.right_eye = 1
b.left_eye = 1
```

```python
import random
import time
for i in range (5):
    a.left_eye = random.randint(1, 9) 
    a.right_eye = random.randint(1, 9)
    b.left_eye = random.randint(1, 9) 
    b.right_eye = random.randint(1, 9)
    time.sleep(1)
```
