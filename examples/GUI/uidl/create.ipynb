```python
from nanohubuidl.ipywidgets import buildJSX
gingerbread = buildJSX(
    '''<svg class="gingerbread" width="200" height="200" viewBox="-100 -100 200 200">
  <circle class="body" cx="0" cy="-50" r="30" />

  <circle class="eye" cx="-12" cy="-55" r="{left_eye:integer(3)}" onClick="[stateChange(left_eye, 10)]"/>
  <circle class="eye" cx="12" cy="-55" r="{right_eye:integer(3)}" onClick="[propCall(onClickEye, '2')]" />
  <rect class="mouth" x="-10" y="-40" width="20" height="5" rx="2" />

  <line class="limb" x1="{left_arm:integer(-40)}" y1="-10" x2="{right_arm:integer(40)}" y2="-10" />
  <line class="limb" x1="-25" y1="50" x2="0" y2="-15" />
  <line class="limb" x1="25" y1="50" x2="0" y2="-15" />

  <circle class="button" cx="0" cy="-10" r="5" />
  <circle class="button" cx="0" cy="10" r="5" />
</svg>''')

ginger = gingerbread(left_eye=5)
ginger.left_arm=-30
import ipywidgets as w
out = w.Output()
def ClickEye(**kwargs):
    with out:
        print(kwargs)
    ginger.left_arm=-20

ginger.onClickEye(ClickEye)


display(ginger) 
display(out)
```
