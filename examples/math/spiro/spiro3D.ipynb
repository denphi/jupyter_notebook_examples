```python
from numpy import linspace, exp, pi, real, imag
import matplotlib.pyplot as plt
from ipywidgets import interact
```

Here is the basic Spirograph equation

```python
def plot_spiro(n1, n2, n3):
    t = linspace(0, 1, 10000)
    z = exp(1j*2*pi*n1*t) + exp(1j*2*pi*n2*t) + exp(1j*2*pi*n3*t)
    plt.plot(real(z), imag(z))
    plt.show()
```

```python
plot_spiro(13, -7, -3)
```

```python
interact(plot_spiro, n1=(-20,100), n2=(-20,20), n3=(-20,20));
```

```python
import ezvis3d as v3d
import pandas as pd
```

```python
def plot_spiro3d(n1, n2, n3):
    t = linspace(0, 1, 10000)
    z = exp(1j*2*pi*n1*t) + exp(1j*2*pi*n2*t) + exp(1j*2*pi*n3*t)
    return real(z),imag(z), t
```

```python
n1, n2,n3 = 100, -4, -1
x, y, z = plot_spiro3d(n1, n2, n3)
df_data = pd.DataFrame({'x': x, 'y': y, 'z': z})

g = v3d.Vis3d()
g.width = '800px'
g.height = '800px'
g.style = 'line'
g.dataColor = {'stroke': '#3267D2', 'strokeWidth': 5}
g.showPerspective = False
g.keepAspectRatio = True
g.verticalRatio = 1.0
g.plot(df_data)
```

```python
g = v3d.Vis3d()
g.width = '600px'
g.height = '600px'
g.style = 'line'
g.dataColor = {'stroke': '#3267D2', 'strokeWidth': 5}
g.showPerspective = False
g.keepAspectRatio = True
g.verticalRatio = 1.0
g.showShadow=True
def plot_spiro3d(n1, n2, n3):
    t = linspace(0, 1, 10000)
    z = exp(1j*2*pi*n1*t) + exp(1j*2*pi*n2*t) + exp(1j*2*pi*n3*t)
    df_data = pd.DataFrame({'x': real(z), 'y': imag(z), 'z': t})
    return g.plot(df_data)
```

```python
interact(plot_spiro3d, n1=(-20,100), n2=(-20,20), n3=(-20,20));
```
