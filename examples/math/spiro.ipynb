```python
from numpy import linspace, exp, pi, real, imag
import matplotlib.pyplot as plt
```

Here is the basic Spirograph equation.

```python
def plot_spiro(n1, n2, n3):
    t = linspace(0, 1, 1000)
    z = exp(1j*2*pi*n1*t) + exp(1j*2*pi*n2*t) + exp(1j*2*pi*n3*t)
    fig, ax1 = plt.subplots(1)
    ax1.plot(real(z), imag(z))
    ax1.set_aspect('equal')
    plt.show()
```

```python
plot_spiro(13, -7, -3)
```

```python
from ipywidgets import interact
interact(plot_spiro, n1=(-20,20), n2=(-20,20), n3=(-20,20));
```
