## Mini Demo

```python
def plot_bessel(xmax=20, nmin=0, nmax=10, nstep=3):
    from scipy import special
    x = np.linspace(0, xmax, 200)
    f, ax = plt.subplots()
    for n in range(nmin, nmax+1, nstep):
        plt.plot(x, special.jn(n, x), label=r'$J_{%i(x)}$' % n)
    plt.grid()
    plt.legend()
    plt.title('Bessel Functions')

plot_bessel()
```

# Installing it on Your Computer

I recommend using the Anaconda Python distribution.  It works on Windows, Mac, and Linux.

<img src="logo-anaconda.svg" width=400>

[https://www.continuum.io/downloads](https://www.continuum.io/downloads)

Update your distribution with

    conda update --all

I currently recommend using Python 2.7 because some scientific packages have not been updated to work with Python 3.

Add new packages with

    conda install packagename

And if the package is unavailable,

    pip install packagename

# Running the notebook on Your Computer

You simply type "ipython notebook" or "jupyter notebook".

The ipython server will start in your current directory and a browser will open with a directory.
